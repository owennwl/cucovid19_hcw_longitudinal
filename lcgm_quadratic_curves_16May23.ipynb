```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time points
time_points = np.linspace(0, 18, num=100)

# Define the coefficients for the quadratic curves
intercepts = {'PHQ9': 4.51, 'GAD7': 3.71, 'PCL5': 3.01}
slopes = {'PHQ9': 0.01, 'GAD7': -0.06, 'PCL5': -0.08}
quadratic_terms = {'PHQ9': 0.001, 'GAD7': 0.004, 'PCL5': 0.004}

# Define the colors for the curves
colors = {'PHQ9': 'blue', 'GAD7': 'green', 'PCL5': 'red'}

# Plot the curves
for measure, color in colors.items():
    curve_values = (intercepts[measure] +
                    slopes[measure] * time_points +
                    quadratic_terms[measure] * time_points**2)
    plt.plot(time_points, curve_values, label=measure, color=color)

# Add labels and a legend
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend()

# Show the plot
plt.show()
```
